* Publish your SDK on the [PyPI Test Repository](https://test.pypi.org/) to learn the process of deploying packages. You're more than halfway there - continue from the [Generating distribution archives](https://oreil.ly/IaQad) on the Python packaging tutorial.

You have all the tools you need installed already. Because you are using `uv`, you can skip installing `build` and `twine` entirely. `uv` has built-in commands that replace them.

Here is how to translate that tutorial into "the uv way."

### 1\. Go to your SDK Directory

You need to be in the folder where the SDK's `pyproject.toml` lives, not your main project root.

```bash
cd ~/2025/api-book-part-one/chapter7/sdk
```

### 2\. Update Your Metadata (Crucial Step)

Open the `pyproject.toml` file in this directory. You **must** change the name of the package. `swcpy` is likely already taken on PyPI.

Change:
`name = "swcpy"`
To:
`name = "swcpy-doolingdavidrs21"` (or any unique name)

### 3\. Build the Package

Instead of `python3 -m build`, simply run:

```bash
uv build
```

**What this does:** It creates a `dist/` folder containing your `.whl` (wheel) and `.tar.gz` (source) files, just like the tutorial requires.

### 4\. Get Your Token

1.  Go to [TestPyPI Register](https://test.pypi.org/account/register/) and create an account.
2.  Verify your email.
3.  Go to **API Tokens**.
4.  Create a new token. Set the Scope to **"Entire Account"** (since you don't have any projects yet).
5.  **Copy the token.** It will start with `pypi-`.

### 5\. Publish to TestPyPI

Instead of installing `twine` and running `python3 -m twine upload...`, use the `uv publish` command.

Run this command (replace `<YOUR_TOKEN>` with the actual token you just copied):

```bash
uv publish --token <YOUR_TOKEN> --publish-url https://test.pypi.org/legacy/
```

**What this does:**

  * `--token`: Authenticates you without needing a username/password.
  * `--publish-url`: Tells `uv` to send the code to **TestPyPI** instead of the real PyPI production server.

### Quick Summary of the "Translation"

| Tutorial Command | uv Command |
| :--- | :--- |
| `pip install build` | *(Not needed)* |
| `python3 -m build` | **`uv build`** |
| `pip install twine` | *(Not needed)* |
| `python3 -m twine upload ...` | **`uv publish ...`** |
